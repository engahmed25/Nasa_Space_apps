```python
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
import re
```

```python
page = 0
home__name=[]
home__loc=[]
time__share=[]
home__area=[]
home__price=[]
```

```python
for page in range(0,100):
    result = requests.get(f"https://homele.com/ar/properties?purpose=FS&city_id=3&search_term=all&agency_id=any&area_from=0&area_to=0&price_from=0&price_to=0&bedrooms=&bathrooms=&advance_filter=&rental_frequency=&sort_by=newest&completion_status=all&page={page}")
    src = result.content
    soup = BeautifulSoup(src , "lxml")
    home_name = soup.find_all("h5" ,{"class":"title text-truncate"} )
    home_loc = soup.find_all("div" ,{"class":"text-truncate address"} )
    home_price = soup.find_all("div" ,{"class":"text-right"} )
    home_area = soup.find_all("div" ,{"class":"item item-area"} )
    time_share = soup.find_all("div",{"class":"posted-date"})
    for i in range(len(home_name)):
        home__name.append(home_name[i].text)
    for i in range(len(home_loc)):
        z = home_loc[i].text
        step_1=re.findall('[A-Za-zأ-ي0-9._,"]+',z)

        f=(' ').join(step_1)
        home__loc.append(f)
        
    for i in range(len(time_share)):
    
        zzzzz = time_share[i].text
        step_1=re.findall('[A-Za-z0-9._,"]+',zzzzz)

        fff=('/').join(step_1)
        time__share.append(fff)
    
    for i in range(len(home_name)):
        z = home_area[i].text
        step_1=re.findall('[A-Za-zأ-ي0-9._,$"]+',z)

        f=(' ').join(step_1)
        home__area.append(f)
    for i in range(len(home_name)):
        z = home_price[i].text
        step_1=re.findall('[A-Za-zأ-ي0-9._,$"]+',z)

        f=(' ').join(step_1)
        home__price.append(f)     
```

```python
import pandas as pd

data = {
  "name": home__name,
  "loc": home__loc,
    "area": home__area,
    "price" : home__price,
    "date" :time__share
}

df = pd.DataFrame(data)

df
```
